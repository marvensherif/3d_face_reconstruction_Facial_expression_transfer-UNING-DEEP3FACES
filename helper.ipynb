```python
from mtcnn import MTCNN
from PIL import Image
import numpy as np

# Create a detector instance specifying the device
detector = MTCNN(device="cpu")

# Load an image using PIL
image = Image.open("your image path")

# Convert image to RGB to ensure compatibility
image = image.convert('RGB')

# Detect faces in the image
results = detector.detect_faces(np.array(image))

# Process each face found and print keypoints along with confidence
for idx, result in enumerate(results):
    confidence = result['confidence']
    keypoints = result['keypoints']
    
    # Only print confidence if more than one face detected
    confidence_output = f"Confidence: {confidence:.2f}\n" if len(results) > 1 else ""
    
    # Prepare to print each keypoint coordinate on a new line without labels and brackets
    keypoints_output = (
        f"{confidence_output}"
        f"{keypoints['left_eye'][0]} {keypoints['left_eye'][1]}\n"
        f"{keypoints['right_eye'][0]} {keypoints['right_eye'][1]}\n"
        f"{keypoints['nose'][0]} {keypoints['nose'][1]}\n"
        f"{keypoints['mouth_left'][0]} {keypoints['mouth_left'][1]}\n"
        f"{keypoints['mouth_right'][0]} {keypoints['mouth_right'][1]}\n"
    )
    
    print(keypoints_output)
```
